## Simple To-Do List Manager


### Build a program that manages a to-do list with add, view, and remove functionality.
#### Requirements:
##### Display a menu with options: Add task, View tasks, Remove task, Exit
- Add task: ask for task description and add it to a list 
- View tasks: display all tasks with numbers 
- Remove task: ask for task number and remove it from the list 
- Loop until the user chooses Exit 
- Use functions for each menu option 
- Handle invalid input (e.g., removing a task that doesn't exist)


```python
tasks = []

while True:
    print("\n === Task Manager ====")
    print("Select 1 to add a new task")
    print("Select 2 to view all tasks")
    print("Select 3 to remove a task")
    print("Select 4 to exit the programme")

    choice = input("Enter your choice")

    if choice == "1":
        title = input("Enter the title of your task:")
        descr = input("Decsribe your task in detail:") 
        tasks.append({"title":title, "description":descr})
        print(f"Your new task, {title} has been added to your to-do list")

    elif choice == "2":
        for i, task in enumerate(tasks, 1):
            print(f"{i}. {task['title']} - {task['description']}")
        
    elif choice == "3":
        delete = input("Enter the title of the task you want to delete:")

        for task in tasks:
            if task['title'] == delete:
                tasks.remove(task)
                print(f"{delete} has been removed from your to-do list")
                break
            else:
                print(f"'{delete}' was not found in your task list.")

    elif choice == "4":
        print("Thanks and Goodbye!")
        break

    else:
        print("Invalid choice, please select 1,2,3 or 4.")
```
