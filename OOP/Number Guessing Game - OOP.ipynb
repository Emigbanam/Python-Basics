## Number Guessing Game Build a game where the computer picks a random number and the user tries to guess it.

### Requirements:
#### Generate a random number between 1 and 100
- Ask the user to guess the number 
- Tell them if their guess is too high, too low, or correct 
- Keep track of the number of guesses 
- When they guess correctly, display the number of attempts 
- Ask if they want to play again 
- Use at least 3 functions (e.g., generate_number, get_guess, check_guess)

```python
import random

class NumberGuesser:

    def __init__(self):
        self.attempts = 0
        self.random_num = None

    def generate(self):
        self.random_num = random.randint(1,100)
        return self.random_num
    
    def get_guess(self):
        guess = int(input("Guess the number"))
        return guess
    
    def check_guess(self, guess):
        if guess > self.random_num:
            print("Too high, try something lower")
        elif guess < self.random_num:
            print("Too low, try something higher")
        else:
            print("Correct guess!")
        return guess == self.random_num
    
    def play_again(self):
        choice = input("Would you like to play again?\nyes/no: ")
        if not choice.lower() != "no":
            print("Thank you for playing!")
            return False
        return True



game = NumberGuesser()

print("====Welcome to Guesser Game====")
print("How to play. \nI will select a number and you will have to guess it")
print("Let's begin!")


while True:
    game.generate()
    game.attempts = 0

    while True:
        guess = game.get_guess()
        game.attempts += 1

        if game.check_guess(guess):
            print(f"You got it in {game.attempts} attempt(s)!")
            break

    if not game.play_again():
        print("Thank you for playing!")
        break
```
