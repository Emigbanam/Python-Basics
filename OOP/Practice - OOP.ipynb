## Object-Oriented Programming (OOP) — Broken Down Simply

### The Big Idea: What is OOP?
- Before OOP, you wrote code like a recipe — a list of instructions that run one after another. OOP is different. Instead of instructions, you think in objects.
- Real-world parallel: Think of a building. It has properties (name, denomination, location, membership count) and it does things (holds services, registers members, collects offerings). In OOP, you'd model that building as an object.


-----------
### 1. Classes and Objects

- Class — the blueprint or template. Like an architectural drawing.
- Object / Instance — the actual thing built from that blueprint. Like the real building.



Rules to follow when defining a class:

1. Class name must start with a capital letter
2. The `class` keyword must come first
3. The class body must be indented
4. `self` must be the first parameter of every instance method
5. `__init__` must be spelled exactly right if used
6. Class names cannot contain spaces or start with a number
7. An empty class body must have `pass` as a placeholder

````python
#Defining a class

class Building:
    pass

#Creating objects (in this case instances of the Building class)
build1 = Building()
build2 = Building()
````

### 2. Attributes — Giving Objects Properties
##### Attributes are just variables that belong to an object. There are two kinds:
*Class attributes — shared by all objects of that class:*

````python
class Building:
    type = 'Bungalow'

build1 = Building()
build2 = Building()

print(build1.type)
print(build2.type)
````

*Instance attributes — unique to each object:*

````python
build1.colour = 'Green'
build1.location = 'New Benin'

build2.colour = "White"
build2.location = 'Ugbowo'

print(build1.colour)
print(build2.colour)
````

### 3. ```__init__``` — The Constructor
Typing ```build2.colour = ...``` manually each time is tedious. The ```__init__``` method solves this — it runs automatically the moment you create an object, and sets everything up for you.
```__init__``` is short for *initialise.* Think of it as the "setup form" that runs when a new object is born.

````python
class Building:
    def __init__(self, colour, location):
        self.colour = colour
        self.location = location

build1 = Building('Greeny', 'New Bini')
build2 = Building('Bluey', 'Ugbowor')

print(build1.colour, build2.colour)
````

````python
class Players:
    def __init__(self, name, age, country, club, goals):
        self.name=name
        self.age = age
        self.country = country
        self.club=club
        self.goals=goals

player1 = Players('Ronaldo', 41, 'Portugal', 'Al Nassr', 967)
player2 = Players('Messi', 38, 'Argentina', 'Inter Miami', 903)

print(f"The two players in thw world are {player1.name} and {player2.name}.\nEach of them has {player1.goals} and {player2.goals} respectively")
````

````python
class Goats:
    def __init__(self, name, country):
        self.name = name
        self.country = country

goat1 = Goats('Cristiano', 'Portugal')
goat2 = Goats('Messi', 'Argentina')
goat3 = Goats('Pele', 'Brazil')
goat4 = Goats('Maradona', 'Argentina')

print(f"These are the goats of football.\n1.{goat1.name}\n2.{goat2.name}\n3.{goat3.name}\n4.{goat4.name}")
````

### 4. Methods — Giving Objects Behaviour
A method is just a function that belongs to a class. It defines what an object can do.


__A class describes what an object is and what it can do.__
- The _"what it is"_ part is the attributes — the data stored on the object like name, colour, location.
- The _"what it can do"_ part is the methods. A method is simply an action the object can perform.

````python
class BankAccount:
    def __init__(self, name, balance=0):
        self.name = name
        self.balance = balance

    def deposit(self, amount):
        self.balance += amount
        print(f"Your deposited amount is:₦{amount}.\nNew balance is ₦{self.balance} ")

    def withdrawal(self, amount):
        if self.balance < amount:
            print('Insufficient Funds')
        else:
            self.balance = self.balance - amount
            print(f"You have wthdrawn ₦{amount}.\nYour remaining balance is ₦{self.balance} ")
    
    def check_balance(self):
        print(f"Your balance is ₦{self.balance}")
````

````python
my_account = BankAccount('Emig', 999999999)
my_account.withdrawal(40000)
my_account.check_balance()
````

````python
class BankAccount:
    total_accounts = 0

    def __init__(self, owner):
        self.owner = owner
        BankAccount.total_accounts += 1

    @classmethod
    def get_total(cls):
        return cls.total_accounts
    
````

````python
account1 = BankAccount('Badru')
account2 = BankAccount('Oseni')
print(BankAccount.get_total())
````

### 4. Encapsulation — Hiding the Inner Workings
Encapsulation means protecting sensitive data inside an object so nothing outside can accidentally break it.



Python uses naming conventions to indicate access levels:
- **Public:** normal names (`self.name`)
- **Private:** prefix with `__` (`self.__balance`)

----
Think of an ATM. You can deposit and withdraw, but you can't directly reach into the machine and change the balance figure manually. The balance is *hidden*, and you can only interact with it through proper channels (deposit/withdraw buttons).
In Python, you make an attribute private by putting double underscores in front of it ```(__balance):```


````python
class Account:
    def __init__(self, owner, balance):
        self.owner = owner
        self.balance = balance

    def acc_bal(self):
        return self.balance

account1 = Account('BopZaddy', 5000000)

print(account1.acc_bal())
````

````python
class BankAccount:
    def __init__(self, owner, balance):
        self.owner = owner
        self.balance = balance      # public

    def deposit(self, amount):
        self.balance += amount

    def withdraw(self, amount):
        if amount > self.balance:
            print("Insufficient funds")
        else:
            self.balance -= amount

account = BankAccount("Emigbanam", 50000)
account.deposit(10000)
account.withdraw(5000)
print(account.balance)          # 55000

# But anyone can bypass the methods entirely:
account.balance = -9000000      # no error, no warning — just breaks it
print(account.balance)          # -9000000
````

````python
class BankAccount:
    def __init__(self, owner, balance):
        self.owner = owner
        self.__balance = balance    # private — double underscore added

    def deposit(self, amount):
        self.__balance += amount

    def withdraw(self, amount):
        if amount > self.__balance:
            print("Insufficient funds")
        else:
            self.__balance -= amount

    def get_balance(self):          # needed because __balance is now hidden
        return self.__balance

account = BankAccount("Emigbanam", 50000)
account.deposit(10000)
account.withdraw(5000)
print(account.get_balance())    # 55000

# Try to bypass the methods:
account.__balance = -9000000    # appears to work but doesn't touch the real balance
print(account.get_balance())    # still 55000 — protected
````

````python
class BankAccount:
    def __init__(self, owner, balance):
        self.owner = owner
        self.balance = balance

    def get_balance(self):
        return self.balance

account = BankAccount("Emigbanam", 50000)
account.balance = -9000000
print(account.get_balance())  # still 55000 — not affected
````

````python
class BankAccount:
    def __init__(self, owner, balance):
        self.owner = owner
        self.__balance = balance

    def get_balance(self):
        return self.__balance

account = BankAccount("Emigbanam", 50000)
account.__balance = -9000000
print(account.get_balance())  # still 55000 — not affected
````

````python
class BankAccount:
    def __init__(self, owner, balance):
        self.owner = owner
        self.__balance = balance

    def get_balance(self):
        return self.__balance
    
    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
        
````

### 6. Inheritance — Building on What Already Exists
Inheritance lets you create a new class that inherits all the features of an existing one, and then adds or changes what it needs to.

-----
__Real-world parallel:__ A Methodist church and a Baptist church are both _churches_ — they share many things (they hold services, have members, collect offerings). But they also have differences. In OOP, you'd have a parent class `Church` and child classes `Methodist` and `Baptist`.

````python
class Animal:
    def __init__(self,name):
        self.name = name

    def speak(self):
        print(f"{self.name} makes a sound")
````

````python
class Dog(Animal):
    def speak(self):
        print(f"{self.name} says woof!")


class Cat(Animal):
    def speak(self):
        print(f"{self.name} says Meow!")
````

````python
dog = Dog('Bingo')
cat = Cat("Whiskers")
dog.speak()
cat.speak()
````

### 6.2 Extending the Parent Class

### `super()` is called a **built-in function.**

__What it does:__
It gives you access to the parent class from inside the child class. When you call `super().__init__(name)` inside `Dog`, you are saying — _"run the `__init__` that belongs to my parent class `Animal`, and pass `name` into it."_






----

__Why is it needed here?__
`Dog` has its own `__init__` that needs two things — `name` and `breed`. But `Animal` already has an `__init__` that handles `name`. Rather than rewriting that logic again inside `Dog`, you just call `super()` to let `Animal` handle `name`, then `Dog` handles the rest.






----



````python
#Parent Class

class Animal:
    def __init__(self,name):
        self.name = name

    def speak(self):
        print(f"{self.name} makes a sound")
````

````python
#Without super():

class Dog(Animal):
    def __init__(self, name, breed):
        self.name = name   # rewriting Animal's work manually
        self.breed = breed
````

````python
#With super():

class Dog(Animal):
    def __init__(self, name, breed):
        super().__init__(name)  # Animal handles name
        self.breed = breed      # Dog just adds breed
````

````python
class Animal:
    def __init__(self, name, age, weight, height, colour, diet, habitat, lifespan, speed, sound):
        self.name = name
        self.age = age
        self.weight = weight
        self.height = height
        self.colour = colour
        self.diet = diet
        self.habitat = habitat
        self.lifespan = lifespan
        self.speed = speed
        self.sound = sound
````

````python
#Without Super()
class Dog(Animal):
    def __init__(self, name, age, weight, height, colour, diet, habitat, lifespan, speed, sound, breed):
        self.name = name
        self.age = age
        self.weight = weight
        self.height = height
        self.colour = colour
        self.diet = diet
        self.habitat = habitat
        self.lifespan = lifespan
        self.speed = speed
        self.sound = sound
        self.breed = breed
````

````python
#With Super()
class Dog(Animal):
    def __init__(self, name, age, weight, height, colour, diet, habitat, lifespan, speed, sound, breed):
        super().__init__(name, age, weight, height, colour, diet, habitat, lifespan, speed, sound)
        self.breed = breed
````

### 7. Magic Methods — Customising How Objects Behave

**Magic methods** (also called __dunder methods__ because they have __`double underscores`__ on both sides) let you customise how Python treats your objects.

-----

| Method | What it controls |
|---|---|
| `__init__` | What happens when an object is created |
| `__str__` | What `print(object)` shows |
| `__repr__` | What the developer sees in the console |
| `__len__` | What `len(object)` returns |
| `__eq__` | What `object1 == object2` means |

````python
class Book:
    def __init__(self, title, pages):
        self.title = title
        self.pages = pages
    
    def __len__(self):
        return self.pages
    
    def __eq__(self, other):
        return self.pages == other.pages

book1 = Book("Python Guide", 300)
book2 = Book("ML Basics", 300)

print(len(book1))
print(book1 == book2)
````

#### `__init__` — runs when an object is created
It's the setup method that runs automatically the moment you create an object.

````python
class Dog:
    def __init__(self,name,age):
        self.name = name
        self.age = age

dog = Dog('Bobby', 3)
print(f"{dog.name} is {dog.age} years old")
````

#### `__str__` — controls what `print(object)` shows

Without `__str__`, printing an object gives you something ugly and useless like `<__main__.Dog object at 0x7f3c1>`. `__str__` lets you decide what a human-friendly description of the object looks like.

````python
#Without __str__

class Dog:
    def __init__(self,name,age):
        self.name = name
        self.age = age

dog = Dog('Bobby', 3)

print(dog)
````

````python
#With __str__

class Dog:
    def __init__(self,name,age):
        self.name = name
        self.age = age

    def __str__(self):
        return f"{self.name}, {self.age} years old"

dog = Dog('Bobby', 3)

print(dog)
````

#### `__repr__` — controls what the developer sees in the console

While `__str__` is for regular users, `__repr__` is for developers. It's meant to show a technical, precise description of the object — ideally something you could copy and paste back into Python to recreate the object.

````python
class Dog:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def __str__(self):
        return f"{self.name}, {self.age} years old"
    
    def __repr__ (self):
        return f"Dog('{self.name}', {self.age})"


dog = Dog("Buddy", 3)
print(dog)          # Buddy, 3 years old  — uses __str__
print(repr(dog))    # Dog('Buddy', 3)     — uses __repr__
````

#### `__len__` — controls what len(object) returns


Normally `len()` works on things like lists and strings. `__len__` lets you define what "length" means for your own custom object.

````python
class Playlist:
    def __init__(self, name, songs):
        self.name = name
        self.songs = songs
    def __len__(self):
        return len(self.songs)
    

my_playlist = Playlist('Afrobeat Mix', ['song1', 'song2', 'song3', 'song4'])
print(len(my_playlist))
````

#### `__eq__` — controls what == means between two objects

By default, `==` between two objects just checks if they are the exact same object in memory — not if they contain the same data. `__eq__` lets you define what "equal" actually means for your objects.

````python
class Dog:
    def __init__ (self, name, age):
        self.name = name
        self.age = age

    def __eq__ (self, other):
        return self.name == other.name and self.age == other.age
    
dog1 = Dog('Buddy', 3)
dog2 = Dog('Buddy', 3)
dog3 = Dog('Luna', 5)

print(dog1 == dog2)     
print(dog1 == dog3)     
````
