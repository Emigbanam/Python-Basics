## Simple To-Do List Manager (Using Classes)


### Build a program that manages a to-do list with add, view, and remove functionality.
#### Requirements:
##### Display a menu with options: Add task, View tasks, Remove task, Exit
- Add task: ask for task description and add it to a list 
- View tasks: display all tasks with numbers 
- Remove task: ask for task number and remove it from the list 
- Loop until the user chooses Exit 
- Use functions for each menu option 
- Handle invalid input (e.g., removing a task that doesn't exist)


```python
class Task:
    def __init__(self, title, description):
        self.title = title
        self.description = description


class TaskMngr:
    def __init__(self):
        self.tasks = []

    def add_task(self, title, description):
        task = Task(title, description)
        self.tasks.append(task)
        print(f"New task added to your To-do List")

    def view_task(self):
        if self.tasks == []:
            print("No task found")
        else:
            for i, task in enumerate(self.tasks, 1):
                print(f"{i}. {task.title} - {task.description}")

    def remove_task(self, title):
        for task in self.tasks:
            if task.title.lower() == title.lower():
                self.tasks.remove(task)
                print(f"{task.title} has been successfully deleted")
                return
        else:
            print("Task not found")


tasks = TaskMngr()

while True:
    print("\n === Task Manager ====")
    print("Select 1 to add a new task")
    print("Select 2 to view all tasks")
    print("Select 3 to remove a task")
    print("Select 4 to exit the programme")

    choice = input("Enter your choice")

    if choice == "1":
        title = input("Enter the title of your task:")
        description = input("Decsribe your task in detail:") 
        tasks.add_task(title, description)
        

    elif choice == "2":
        tasks.view_tasks()
        
    elif choice == "3":
        title = input("Enter the title of the task you want to delete:")
        tasks.remove_task(title)

    elif choice == "4":
        print("Thanks and Goodbye!")
        break

    else:
        print("Invalid choice, please select 1,2,3 or 4.")
```
