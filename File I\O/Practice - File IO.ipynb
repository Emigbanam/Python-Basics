### 🔹 Why File I/O Matters (Big Picture First)

Normally, when you run a Python program, everything disappears when it ends.

It’s like writing something on a whiteboard and wiping it immediately.

File I/O fixes that.

Now your program can:

Save data → like saving contacts to your phone
Load data → like opening WhatsApp and seeing old chats
Share data → like exporting a report

A banking app saves your balance in a file (or database). If it didn’t, your money would reset every time you open the app 😅

#### 🔹 Section 1: Reading Text Files
##### 1.1 Basic File Reading

This is simply:
👉 “Open a file and read everything inside it”

💡 Example:
You open a notebook and read everything written in it.

````python
file = open('traaqa.txt', 'r')
content = file.read()
print(content)
file.close()
````

`'r'` means read mode
`read()` means read everything

⚠️ Important:
You must __close__ the file, like closing a book after reading.




🔹 So why must you close it?
1. To free up system resources
2. To make sure your data is actually saved

````python
# readline() — reads one line at a time
file = open('traaqa.txt', 'r')
line1 = file.readline()
line2 = file.readline()
print(f"Line 1: {line1}")
print(f"Line 2: {line2}")
file.close()
````

##### 1.2 Reading Line by Line

````python
file = open('traaqa.txt', 'r')
print(file.readline())
print(file.readline())
file.close()
````

##### 1.3 Iterating Through Lines

````python
# readlines() — reads all lines into a list
file = open('traaqa.txt', 'r')
lines = file.readlines()
for line in lines:
    print(line.strip())  # strip() removes \n
file.close()
````

````python
with open('traaqa.txt', 'r') as traaqa:
    for line in  traaqa.readlines():
        print(line)
````

#### 🔹 Section 2: Writing Text Files

##### 2.1 Basic Writing
`open('output.txt', 'w')`

👉 `'w'` means:

- Create a new file OR
- Delete everything inside and start fresh

💡 Real-life:
Like opening a notebook and tearing out all pages before writing again 😬

````python
# Write to a file (overwrites if exists)
file = open('output.txt', 'w')
file.write('Hello, File!\n')
file.write('This is a new line.\n')
file.close()


file = open('output.txt', 'r')
print(file.read())
file.close()
````

````python
with open('write_test.txt', 'a') as filex:
    filex.write('\nIt is our dream that many people will enjoy using our app.')

with open('write_test.txt', 'r') as filex:
    for line in filex.readlines():
        print(line)    
````

##### 2.2 Appending

````python
with open('traaqa.txt', 'a') as filex:
    filex.write('\n We hope you enjoy using our app.')

with open('traaqa.txt', 'r') as filey:
    content = filey.read()
    print(content)
````

### 🔹 Section 3: File Modes

These are just different ways to open a file.

Think of it like opening a door:


| Mode | Description |
|------|-------------|
| `'r'` | Read (default). File must exist: *Just look (read only)* |
| `'w'` | Write. Creates file or overwrites: *Start fresh (erase + write)* |
| `'a'` | Append. Creates file or adds to end: *Add more*|
| `'r+'` | Read and write. File must exist: *Read + edit*|
| `'w+'` | Read and write. Creates or overwrites: *Reset + read + write*|
| `'rb'` | Read binary. |
| `'wb'` | Write binary. |

### Section 4: The with Statement
`with` is a way of saying:

_“Use this resource for a short time, and clean it up automatically when you're done.”_

That “resource” can be:

- a file
- a network connection
- a database
- etc.


#### 🔹 Compare the two approaches
##### ❌ Without with
- `file = open("data.txt", "r")`
- `content = file.read()`
- `file.close()`

*This works… but only if everything goes perfectly.*

🔹 The hidden problem

What if something goes wrong?

- `file = open("data.txt", "r")`
- `content = file.read()`

- `# 💥 error happens here`
- `print(10 / 0)`

- `file.close()`  # ❌ this never runs

👉 The file is left open

##### ✅ With with
- `with open("data.txt", "r") as file:`
-     `content = file.read()`

##### Even if an error happens inside:

👉 Python automatically closes the file for you

## Section 5.2 — Building File Paths That Work on Any Computer

Imagine you're giving someone directions to your house. If you live in Lagos and you tell them *"turn left at the big mango tree"*, those directions only make sense locally — someone in Abuja has no idea what mango tree you're talking about. File paths have the same problem.

Windows computers use backslashes to separate folders: `data\files\output.txt`. Mac and Linux use forward slashes: `data/files/output.txt`. If you type a path manually in your code, it works on your computer and breaks on someone else's.

---

### The Problem — A File Buried 5 Folders Deep

Imagine you're building a school management system and a student report is stored here:

```
school → academics → year2 → term1 → reports → student_report.txt
```

A beginner might type the path directly:

```python
# Works on Mac/Linux, breaks on Windows
path = 'school/academics/year2/term1/reports/student_report.txt'

# Works on Windows, breaks on Mac/Linux
path = 'school\\academics\\year2\\term1\\reports\\student_report.txt'
```

Both are fragile. You'd have to rewrite the path every time you shared your code with someone on a different computer.

---

### The Solution — `os.path.join()`

Instead of typing the slash yourself, you hand Python the folder names as separate pieces and let it figure out the correct separator for whatever computer the code is running on:

```python
import os

path = os.path.join('school', 'academics', 'year2', 'term1', 'reports', 'student_report.txt')

print(path)
# On Mac/Linux →  school/academics/year2/term1/reports/student_report.txt
# On Windows   →  school\academics\year2\term1\reports\student_report.txt

# Use it to open the file
with open(path, 'r') as f:
    print(f.read())
```

Think of `os.path.join()` like a GPS app. You don't tell the GPS *"turn left, then right"* — you just give it the destination pieces and it figures out the correct route for your current location. You just give Python the folder names and it builds the correct path for whichever operating system is running.

---

### Why is this better? Three reasons

**1. It works everywhere, automatically.**
Write the code once. Your classmate on Windows, your lecturer on a Mac, and a server running Linux will all get the correct path without anyone changing a single character.

**2. It's easier to read and modify.**
Each folder name is clearly separated by a comma. If you need to add or remove a folder, you just add or remove one item — no hunting for slashes inside a long string.

**3. It handles edge cases for you.**
If a folder name accidentally has a trailing slash, `os.path.join()` is smart enough not to double it up. If you built the path manually, you'd have to check for that yourself.

---

### Bonus Tool — `os.path.abspath()`

A relative path like `'student_report.txt'` is a shortcut — it only makes sense if you know which folder Python is currently running from. `abspath()` converts that shortcut into the full, unambiguous address:

```python
abs_path = os.path.abspath('student_report.txt')
print(abs_path)
# Mac/Linux →  /home/alice/school/academics/year2/term1/reports/student_report.txt
# Windows   →  C:\Users\Alice\school\academics\year2\term1\reports\student_report.txt
```

This is like the difference between *"I live two streets down"* (only useful if you know where the conversation is happening) versus *"14 Adeola Street, Victoria Island, Lagos"* — unambiguous from anywhere in the world.

---

> **The golden rule:** never type slashes in your paths manually. Always use `os.path.join()` to build paths, and your code will run correctly on any computer without modification.

### 🔹 Section 6: Working with CSV Files

### What is a CSV file?

A **CSV (Comma-Separated Values)** file is a simple spreadsheet stored as plain text. Each row is a line, and each column is separated by a comma. Think of it as an Excel spreadsheet you can open in Notepad.

Example of what a CSV looks like inside:
```
Name,Age,City
Alice,28,NYC
Bob,32,LA
```

````python
import csv

with open('students.csv', 'r') as students:
    reader = csv.reader(students)
    for row in reader:
        print(row) 
````

##### 6.2 Reading CSV as Dictionaries

````python
with open('students.csv', 'r') as stdnt:
    for row in csv.DictReader(stdnt):
        print(row)
````

##### Why would you want to read your csv files as dictionaries?

Great question! It's all about **convenience when accessing specific values.**

When you read as a plain list, each row comes back like this:

```python
['Alice', '28', 'NYC']
```

To get Alice's city, you'd have to write `row[2]` — you have to **remember that index 2 means City**. That's fine for 3 columns, but imagine a spreadsheet with 20 columns. Remembering that index 14 means "postal code" becomes a nightmare.

When you read as a dictionary, the same row comes back like this:

```python
{'Name': 'Alice', 'Age': '28', 'City': 'NYC'}
```

Now you just write `row['City']` — **no memorizing positions**, the column name tells you exactly what you're getting.

---

A real-life comparison: imagine you ordered a package and it arrives in a plain unlabelled box with items just thrown inside. That's the list — you have to dig around and count positions to find what you want. The dictionary version is the same box but every item has a sticky label on it. Much easier to find what you need.

---

It becomes especially useful when your code needs to do something specific, like:

```python
# Only print people who are older than 30
for row in reader:
    if int(row['Age']) > 30:
        print(row['Name'])
```

Try doing that clearly with index numbers — `row[1]` and `row[0]` — and your code becomes much harder to read and understand at a glance.

````python
with open('students.csv', 'r') as stdnt:
    for row in csv.DictReader(stdnt):
        print(f"{row['Name']} is {row['Age']} years old")
````

#### 6.3 Writing CSV Files
You can write multiple rows at once with `writerows()` (notice the **s**):

````python
with open('data.csv', 'w') as file1:
    writer = csv.writer(file1)
    writer.writerow(['Name', 'Age', 'City'])  #This row will bw the header
    writer.writerow(['Alex', 31, 'Lag'])
    writer.writerow(['Barb', 26, 'PH'])
    writer.writerow(['Chazz', 23, 'ABJ'])
````

Alternatively you can `write` the csv file via a list

````python
data = [
    ['Name', 'Age', 'City'],
    ['Jane', 32, 'Abj'],
    ['Wale', 30, 'Lag'],
    ['Jummy', 25, 'PH']
]


with open('members.csv', 'w', newline='') as filey:
    writerx =  csv.writer(filey)
    writerx.writerows(data)


with open('members.csv', 'r') as readex:
    readey = csv.reader(readex)
    for row in readey:
        print(row)
````

````python
with open('members.csv', 'r') as readex:
    readey = csv.DictReader(readex)
    for row in readey:
        print(f"My name is {row['Name']}, I'm {row['Age']} years old, and I reside in {row['City']} city.\nI am excited to meet you!")
````

#### Section 7: Working with JSON Files
##### 7.1 What Is JSON?
**JSON (JavaScript Object Notation)** is a way to store structured, labeled data as text — like a very organized checklist or a form with named fields. It's the most popular format for saving structured data and sharing data between apps. In simple terms, it is a text format for storing structured data.

- Think of JSON as a *labeled storage box:* each compartment has a name (like "name", "age") and a value stored inside it. It's human-readable AND machine-readable.
- *Why JSON?* Unlike CSV, JSON can store nested or complex data (lists inside dictionaries, etc.), making it perfect for app settings, API responses, and configuration files.
- Human-readable
- Language-independent
- Widely used for APIs and config files

##### 7.2 Writing & Reading JSON

````python
import json


data = {
    'name':'Alice',
    'age':28,
    'city':'NYC',
    'hobbies':['reading', 'hiking']
}


with open('person.json', 'w') as fileq:
    json.dump(data, fileq, indent=2)


with open('person.json', 'r') as fileq1:
    data1 = json.load(fileq1)

print(data1)
print(f"Name: {data1['name']}")       # Alice
print(f"Hobbies: {data1['hobbies']}") # ['reading', 'hiking']
````

##### 7.4 Working with Lists of Objects
You can also store multiple records (like a list of people) in a single JSON file:

````python
pipo = [
    {'name': 'Alice','age':28},
    {'name':'Bob','age':32},
    {'name':'Chazz','age':25}
]



with open('pipo.json', 'w') as x:
    json.dump(pipo, x, indent=2)


with open('pipo.json', 'r') as y:
    loaded = json.load(y)  


for person in loaded:
    print(f"{person['name']} is {person['age']} years old")
````
