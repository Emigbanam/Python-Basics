## Building a Menu Interface On Python

##### You’re essentially learning how to control how users interact with your program. Think of a menu interface like the options you see on an ATM or a mobile app—“Press 1 to withdraw, 2 to check balance.” In Python, there are several ways to build that experience, and each one fits a different level of complexity.

Let’s walk through them gradually, from the simplest to the more advanced, using a **contact manager** example (add, view, search, delete, exit).

----
### 1. Simple `if/elif`  Menu
##### What it is (in simple terms)


The simplest possible menu. Python reads your choice once, does something, then stops. Like a vending machine that dispenses one item and turns off. This is the most basic way to handle user choices. You ask the user for input once, and then use `if`, `elif`, and `else` to decide what to do.

It’s like asking:




“If the user says 1 → do this, if 2 → do that…”

```python
contacts = []

print ("1. Add Number\n2. View Contacts\n3. Exit")

choice = input('Enter your choice')

if choice == '1':
    name = input('Enter contact name')
    contacts.append(name)
    print(f"{name} has been added to your contacts")

elif choice == '2':
    if contacts:
        for contact in contacts:
            print(f"- {contact}")
        else:
            print('No contact has beeen saved')

elif choice == '3':
    print('Goodbye')

else:
    print('Invalid Choice')
```

### 2. `while` True Loop Menu
##### What it is

This builds on the first approach but keeps the program running until the user chooses to exit. Imagine the vending machine resets itself after every purchase and keeps asking "what do you want?" until you press EXIT. That's a while True loop — it keeps running the menu forever, until the user chooses to quit.

Think of it like an ATM that keeps showing options until you press “Cancel”.

```python
contacts = []

while True:
    print ("1. Add Number\n2. View Contacts\n3. Delete Contcts\n4. Exit")

    choice = input('Enter your choice')

    if choice == '1':
        name = input("Enter contact name:")
        contacts.append(name)
        print(f"{name} has been added to your contact list")

    elif choice == '2':
        if contacts:
            for contact in contacts:
                print(f" - {contact}")
        else:
            print('No saved contacts found')

    elif choice == '3':
        name = input('Enter the name of the contct you want to delete:')
        if name in contacts:
            contacts.remove(name)
            print(f"{name} has been deleted from your contact list")
        else:
            print(f'{name} was not found on your contact list')

    elif choice == '4':
        print('Thanks and goodbye!')
        break

    else:
        print('Invlid choice')


```

### 3. Dictionary-Based Menu
What it is

Instead of using many `if/elif`, you create a "lookup table" (a dictionary) that maps each menu number to a function. Like a hotel switchboard — you press a number, and it connects you directly to the right department.

A dictionary in Python is like a lookup table:

Key → Value
Example: `"1" → function_to_add_contact`

So instead of asking “if this, if that”, you just look up what to run.

This is smarter and scales better — adding a new option is just one line.
